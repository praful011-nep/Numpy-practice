```python
sensors=np.array([
    [12.1,14.3,15.2],
    [10.5,11.2,12.6],
    [14.1,16.4,18.3],
    [18.9,19.1,20.3]
])
```

```python
specific_reading=sensors[2,1]
print(specific_reading)
```

```python
top_sensors=sensors[0:2,:]
print(top_sensors)
```

```python
final_readings=sensors[:, 2]
print(final_readings)
```

```python
#Reshaping
flat_temps=np.array([22,24,25,23,27,24,23,22,25,26,24,25])
structured_temps=flat_temps.reshape(4,3)
print(structured_temps)
```

```python
fahrenheit_temps=(structured_temps*9/5)+32
print("\nTemperatures in Fahrenheit:\n", fahrenheit_temps)
```

```python
scores=np.array([
    [85,90,78,92],
    [70,80,75,85],
    [85,92,88,96]
])
print("Overall highest score in the class:",np.max(scores))
print("Overall average score:",round(np.mean(scores)))
```

```python
subject_averages=np.mean(scores,axis=0).round(2)
print("Average score for each subjects:",subject_averages)
```

```python
student_avereges=np.mean(scores,axis=1)
```

```python

import time
size=1_000_000
python_list=list(range(size))
numpy_array=np.arange(size)
```

```python
start_time=time.time()
python_list_result=[x*2 for x in python_list]
python_duration=time.time()-start_time
```

```python
start_time=time.time()
numpy_array_result=numpy_array*2
numpy_duration=time.time()-start_time
print(f"Time taken by standard Python list:{python_duration:.5f} seconds")
print(f"Time taken by numpy python list: {numpy_duration:.5f} seconds")
print(f"Numpy is roughly {round(python_duration/numpy_duration)}x faster")
```
