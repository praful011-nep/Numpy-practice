```python
import numpy as np
```

```python
#Statistics for data analyiis
#Mean, Median, Mode
from scipy import stats #for mode function as not given by numpy
```

```python
rent_prices=np.array([1200,1350,1200,1400,1550,1300,1200,12000])
```

```python
#Calculating central tendency
mean_rent=np.mean(rent_prices)
median_rent=np.median(rent_prices)
```

```python
mode_result=stats.mode(rent_prices,keepdims=True)
print(mode_result)
mode_rent=mode_result.mode[0]
mode_count=mode_result.count[0]
```

```python
print(f"Mean Rent: ${mean_rent:.2f}")
print(f"Median Rent: ${median_rent:.2f}")
print(f"Mode Rent: {mode_rent} (Appears {mode_count} times)")
```

```python
#Varience and Standard Deviation
city_a_temps=np.array([24,25,26,25,25]) #Mild, stable climate
city_b_temps=np.array([10,40,5,45,25]) #Wild, Volatile climate
```

```python
#Variance calculation
var_a=np.var(city_a_temps)
var_b=np.var(city_b_temps)
```

```python
#Calculate Standard Daviation
std_a=np.std(city_a_temps)
std_b=np.std(city_b_temps)
```

```python
print(f"City A: Variance = {var_a:.2f} | Standard Daviation = {std_a:.2f}")
print(f"City B: Variance = {var_b:.2f} | Standard Daviation = {std_b:.2f}")
```

```python
#Correlation and Covariance
marketing_spend=np.array([10,20,30,40,50])
online_sales=np.array([25,45,55,75,95])
complaints=np.array([15,12,10,4,2])
```

```python
#Covariance matrix between spend and sales
cov_matrix=np.cov(marketing_spend, online_sales)
cov_matrix1=np.cov(marketing_spend, complaints)
print(f"Covariance betweens Spend and Sales:\n{cov_matrix}")
```

```python
#calculate correaltion coefficient matrix
corr_matrix_positive=np.corrcoef(marketing_spend,online_sales)
corr_matrix_negative=np.corrcoef(marketing_spend,complaints)
```

```python
#Extracting the specific relationship calues from the matrices
correlation_spend_sales=corr_matrix_positive[0,1]
correlation_spend_complaints=corr_matrix_negative[0,1]
print(f"Correlation (Spend vs. Sales): {correlation_spend_sales:.4f} (Strong Positive)")
print(f"Correlation (Spend vs. Complaints): {correlation_spend_complaints:.4f} (Strong Negative)")
```
